#### Suppression des lignes avec des valeurs NaN dans une colonne spécifique

```python
def remove_nan_rows(df, column_name):
    """
    Supprime les lignes avec des valeurs NaN dans une colonne spécifique.

    Args:
        df (pandas.DataFrame): Le DataFrame contenant les données.
        column_name (str): Le nom de la colonne à vérifier.

    Returns:
        pandas.DataFrame: Le DataFrame filtré.
    """
    df = df.dropna(subset=[column_name])
    return df


# Utilisation des fonctions pour chaque colonne
df=remove_nan_rows(df, 'Statut - GRTgaz')
df=remove_nan_rows(df, 'Statut - Teréga')
```

#### Pourcentage de valeurs manquantes pour chaque colonne

```python
def handle_missing_values(df):
    """
    Gère les valeurs manquantes dans un DataFrame.

    Si le pourcentage de valeurs manquantes dans une colonne dépasse 55 %,
    les lignes contenant des valeurs manquantes dans cette colonne seront supprimées.
    Sinon, le pourcentage de valeurs manquantes dans cette colonne sera simplement imprimé.

    Args:
        df (pandas.DataFrame): Le DataFrame contenant les données.

    Returns:
        pandas.DataFrame: Le DataFrame après traitement des valeurs manquantes.
    """
    total_rows = len(df)
    for column in df.columns:
        missing_values = df[column].isnull().sum()
        missing_percentage = (missing_values / total_rows) * 100
        print(f"Percentage of missing values in column '{column}': {missing_percentage:.1f}%")

        if missing_percentage > 55:
            print(f"Removing rows with missing values in column '{column}'...")
            df = df.dropna(subset=[column])
            print(f"{len(df)} rows remaining after removing missing values in column '{column}'.")
    return df


df = handle_missing_values(df)
```

#### Visualisation de la distribution des valeurs pour chaque colone du dataframe

```python
def visualize_data(df):
    """
    Visualise la distribution des valeurs pour chaque colonne du DataFrame.

    Args:
        df (pandas.DataFrame): Le DataFrame contenant les données.

    Returns:
        None
    """
    # Itérer à travers chaque colonne du DataFrame
    for column in df.columns:
        # Vérifier si la colonne contient des valeurs numériques
        if df[column].dtype in ['int64', 'float64']:
            # Créer un histogramme pour les valeurs numériques
            plt.figure(figsize=(8, 6))
            sns.histplot(df[column].dropna(), kde=True)
            plt.title(f'Distribution de {column}')
            plt.xlabel(column)
            plt.ylabel('Nombre de cas')
            plt.show()
        else:
            # Créer un graphique à barres pour les valeurs non numériques
            plt.figure(figsize=(8, 6))
            sns.countplot(x=column, data=df)
            plt.title(f'Distribution de {column}')
            plt.xlabel(column)
            plt.ylabel('Nombre de cas')
            plt.xticks(rotation=45)
            plt.show()

# Convertir la colonne Date en format datetime
df['Date'] = pd.to_datetime(df['Date'])

# Fonction pour créer différentes visualisations
def create_visualization(visualization_type):
    if visualization_type == 'Consommation brute de gaz au fil du temps':
        plt.figure(figsize=(12, 6))
        sns.lineplot(x='Date', y='Consommation brute gaz totale (MW PCS 0°C)', data=df)
        plt.title('Consommation brute de gaz au fil du temps')
        plt.xlabel('Date')
        plt.ylabel('Consommation brute de gaz (MW PCS 0°C)')
        plt.xticks(rotation=45)
        plt.show()

    elif visualization_type == 'Distribution de la consommation brute de gaz':
        plt.figure(figsize=(8, 6))
        sns.histplot(df['Consommation brute gaz totale (MW PCS 0°C)'], bins=20, kde=True)
        plt.title('Distribution de la consommation brute de gaz')
        plt.xlabel('Consommation brute de gaz (MW PCS 0°C)')
        plt.ylabel('Nombre d\'occurrences')
        plt.show()

    elif visualization_type == 'Consommation brute de gaz par heure':
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='Heure', y='Consommation brute gaz totale (MW PCS 0°C)', data=df)
        plt.title('Consommation brute de gaz par heure')
        plt.xlabel('Heure')
        plt.ylabel('Consommation brute de gaz (MW PCS 0°C)')
        plt.xticks(rotation=45)
        plt.show()

    elif visualization_type == 'Relation entre la consommation brute de gaz et d\'électricité':
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x='Consommation brute gaz totale (MW PCS 0°C)', y='Consommation brute électricité (MW) - RTE', data=df)
        plt.title('Relation entre la consommation brute de gaz et d\'électricité')
        plt.xlabel('Consommation brute de gaz (MW PCS 0°C)')
        plt.ylabel('Consommation brute d\'électricité (MW)')
        plt.show()

    elif visualization_type == 'Consommation brute de gaz totale par statut de GRTgaz':
        plt.figure(figsize=(8, 6))
        sns.barplot(x='Statut - GRTgaz', y='Consommation brute gaz totale (MW PCS 0°C)', data=df)
        plt.title('Consommation brute de gaz totale par statut de GRTgaz')
        plt.xlabel('Statut - GRTgaz')
        plt.ylabel('Consommation brute de gaz (MW PCS 0°C)')
        plt.show()

    elif visualization_type == 'Consommation brute totale par mois':
        df['Mois'] = df['Date'].dt.month
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Mois', y='Consommation brute totale (MW)', data=df, estimator=sum)
        plt.title('Consommation brute totale par mois')
        plt.xlabel('Mois')
        plt.ylabel('Consommation brute totale (MW)')
        plt.show()

    elif visualization_type == 'Consommation brute de gaz par jour de la semaine':
        plt.figure(figsize=(10, 6))
        df['Day_of_Week'] = pd.to_datetime(df['Date']).dt.dayofweek
        sns.boxplot(x='Day_of_Week', y='Consommation brute gaz (MW PCS 0°C) - GRTgaz', data=df, palette='Set3')
        plt.title('Consommation brute de gaz par jour de la semaine')
        plt.xlabel('Jour de la semaine')
        plt.ylabel('Consommation brute de gaz (MW)')
        plt.xticks(ticks=range(7), labels=['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche'])
        plt.show()

    elif visualization_type == 'Visualiser la distribution des valeurs pour chaque colonne':
        visualize_data(df)



# Liste des types de visualisations disponibles
visualization_types = [
    'Consommation brute de gaz au fil du temps',
    'Distribution de la consommation brute de gaz',
    'Consommation brute de gaz par heure',
    'Relation entre la consommation brute de gaz et d\'électricité',
    'Consommation brute de gaz totale par statut de GRTgaz',
    'Consommation brute totale par mois',
    'Consommation brute de gaz par jour de la semaine',
    'Visualiser la distribution des valeurs pour chaque colonne'
]
# Créer un menu interactif
interact(create_visualization, visualization_type=widgets.Dropdown(options=visualization_types, value=visualization_types[0]))

```

### Impact des Mouvements Sociaux sur la Consommation d'Énergie : Une Analyse autour des Dates de début
---

#### Vérifier l'Intersection des Dates

*Nous allons verifier si les dates des mouvements sociaux ont des correspondances dans le dataset de consommation d'énergie. nous le ferons en examinant l'intersection des dates présentes dans les deux datasets.*

```python
# Dates des mouvements sociaux
social_movement_dates = set(df_mouvement['Date'])

# Dates dans le dataset de consommation d'énergie
energy_dates = set(df_energy['Date'])

# Intersection des dates
common_dates = social_movement_dates.intersection(energy_dates)

print(f"Nombre de jours de mouvements sociaux présents dans les données de consommation : {len(common_dates)}")

# Si l'intersection est non vide, vérifier les valeurs manquantes pour ces dates
if len(common_dates) > 0:
    energy_during_social_movements = df_energy[df_energy['Date'].isin(common_dates)]
    missing_values = energy_during_social_movements['Consommation brute totale (MW)'].isnull().sum()
    print(f"Nombre de valeurs manquantes pour les jours de mouvements sociaux : {missing_values}")
```

*Remarques : le nombre de jours de mouvements sociaux présents dans les données de consommation est de 0, cela signifie qu'il n'y a pas de correspondance directe entre les dates des mouvements sociaux et celles des enregistrements de consommation d'énergie dans notre dataset consomation brute quotidienne.*

#### Étude de l'Influence des Mouvements Sociaux avant et après les Événements (Une Perspective Élargie)

Dans cette section, nous approfondissons notre compréhension de l'influence des mouvements sociaux sur la consommation énergétique. L'objectif est d'examiner non seulement l'effet immédiat de ces événements, mais aussi leur impact avant et après leur survenue. Voici les étapes clés de notre démarche analytique :

- Génération de Périodes Élargies Autour des Jours de Mouvements Sociaux
    + Approche :
            Nous définissons une fenêtre de trois jours avant et après chaque jour de mouvement social enregistré. Cette méthode nous permet d'observer les variations de consommation énergétique associées à l'anticipation et aux retombées des mouvements sociaux.

- Filtrage des Données de Consommation d'Énergie
   + Méthodologie :
            L'ensemble des données sur la consommation d'énergie est ajusté pour se focaliser uniquement sur ces périodes élargies. Cette sélection ciblée affine notre analyse sur les moments pertinents entourant les mouvements sociaux.

- Comparaison des Statistiques de Consommation
    + Analyse :
        Nous calculons et comparons les moyennes de consommation énergétique pendant les périodes liées aux mouvements sociaux avec celles des autres jours. Cela permet de détecter d'éventuels modèles distincts ou variations notables liées à ces événements.

- Interprétation des Résultats


```python
# Générer les périodes autour des jours de mouvements sociaux
days_before_after = 3  # Nombre de jours avant et après les mouvements sociaux à considérer
expanded_dates = []
for date in df_mouvement['Date'].unique():
    current_date = pd.to_datetime(date)
    date_range = pd.date_range(start=current_date - pd.Timedelta(days=days_before_after),
                               end=current_date + pd.Timedelta(days=days_before_after))
    expanded_dates.extend(date_range)

# Suppression des doublons
expanded_dates = list(set(expanded_dates))

# Filtrer les données de consommation d'énergie pour les périodes élargies
df_energy['Date'] = pd.to_datetime(df_energy['Date'])
energy_during_expanded_social_movements = df_energy[df_energy['Date'].isin(expanded_dates)]

# Comparer les statistiques de consommation
mean_consumption_during_expanded = energy_during_expanded_social_movements['Consommation brute totale (MW)'].mean()
mean_consumption_outside_expanded = df_energy[~df_energy['Date'].isin(expanded_dates)]['Consommation brute totale (MW)'].mean()

print(f"Moyenne de la consommation d'énergie pendant et autour des jours de mouvements sociaux : {mean_consumption_during_expanded}")
print(f"Moyenne de la consommation d'énergie en dehors de ces périodes : {mean_consumption_outside_expanded}")
```

*N.B: Cette approche élargie nous offre une perspective plus complète sur la manière dont les mouvements sociaux peuvent influencer les modèles de consommation d'énergie au-delà des effets immédiats, en tenant compte des préparatifs et des retombées de ces événements.*

*Conclusion : Le résultat nan pour la moyenne de la consommation d'énergie pendant et autour des jours de mouvements sociaux indique qu'il n'y a probablement pas de données de consommation d'énergie correspondant aux dates étendues autour des mouvements sociaux, ou que toutes les valeurs de consommation pour ces périodes sont manquantes.*

### Impact des Mouvements Sociaux sur la Consommation d'Énergie : Une Analyse autour des Dates de Fin
---

```python
# Généreration des périodes autour des dates de fin des mouvements sociaux
days_before_after = 3  # Fenêtre temporelle avant et après les dates de fin
expanded_end_dates = []
for end_date in df_mouvement['date_de_fin'].dropna().unique():  # Exclure les valeurs NaT/NaN
    current_end_date = pd.to_datetime(end_date)
    date_range = pd.date_range(start=current_end_date - pd.Timedelta(days=days_before_after),
                               end=current_end_date + pd.Timedelta(days=days_before_after))
    expanded_end_dates.extend(date_range)

# Suppression des doublons
expanded_end_dates = list(set(expanded_end_dates))

# 2. Filtrer les données de consommation d'énergie pour les périodes élargies
energy_during_expanded_end_social_movements = df_energy[df_energy['Date'].isin(expanded_end_dates)]

# 3. Comparer les statistiques de consommation
mean_consumption_during_expanded_end = energy_during_expanded_end_social_movements['Consommation brute totale (MW)'].mean()
mean_consumption_outside_expanded_end = df_energy[~df_energy['Date'].isin(expanded_end_dates)]['Consommation brute totale (MW)'].mean()

print(f"Moyenne de la consommation d'énergie pendant et autour des jours de fin de mouvements sociaux : {mean_consumption_during_expanded_end}")
print(f"Moyenne de la consommation d'énergie en dehors de ces périodes : {mean_consumption_outside_expanded_end}")
```

*Conclusion* : *L'obtention d'une moyenne `nan` pour la consommation d'énergie pendant et autour des jours de fin de mouvements sociaux suggère, une fois de plus, qu'il n'y a pas de correspondance directe entre les dates étendues autour des fins de mouvements sociaux et les enregistrements de consommation d'énergie dans notre dataset, ou que les données correspondantes sont toutes manquantes*.
